#### changing column names to lowercase 

```python
data.columns =  data.columns.str.lower()
data.columns = data.columns.str.replace(' ','_')
data.columns
```

```python
data.head()
```

#### Deriving new columns from existing columns

```python
data['total_discount'] = data['list_price'] * data['discount_percent'] * 0.01 
data['selling_price'] = data['list_price'] - data['total_discount']
data['profit'] = data['cost_price'] - data['selling_price']
data
```

```python
data.dtypes
```

```python
data['order_date'] = pd.to_datetime(data['order_date'], format='%Y-%m-%d')
data['order_date']
```

```python
data.drop(columns=['list_price','discount_percent','cost_price'], inplace=True)
data
```
